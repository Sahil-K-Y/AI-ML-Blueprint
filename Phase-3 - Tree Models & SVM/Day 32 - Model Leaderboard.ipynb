## 🧠 1. Why Systematic Model Comparison?

### The Problem
- Different algorithms have **different strengths**
- A single train/test split gives **unreliable** estimates
- We need a **fair, reproducible** comparison framework

### The Solution: Cross-Validation Leaderboard
1. Use the **same data splits** for all models
2. Report **mean ± std** of scores across folds
3. Measure **training time** alongside accuracy
4. Use **multiple metrics** (accuracy alone can be misleading)

---
## 🧠 2. Cross-Validation Strategies

| Strategy | When to Use |
|----------|-------------|
| **KFold(k=5)** | Balanced datasets, regression |
| **StratifiedKFold(k=5)** | Classification (preserves class distribution) |
| **RepeatedStratifiedKFold** | When you want more stable estimates |
| **LeaveOneOut** | Very small datasets (<100 samples) |
| **TimeSeriesSplit** | Time-dependent data |

### ⚠️ Important Rules
- Always use **StratifiedKFold** for classification
- Use the **same random_state** across all models
- **Never** use test data during model selection

---
## 🧠 3. Algorithm Selection Guide

### When to Use Each Algorithm:

| Algorithm | Best For | Avoid When |
|-----------|----------|------------|
| **Logistic Regression** | Linearly separable, interpretable, fast baseline | Non-linear boundaries |
| **Decision Tree** | Interpretable rules, mixed feature types | Overfitting risk, unstable |
| **SVM (RBF)** | Medium datasets, non-linear, high-dimensional | Very large datasets (slow) |
| **KNN** | Small-medium data, local patterns | High dimensions, large datasets |
| **Naive Bayes** | Text, high-dimensional, fast baseline | Feature interactions matter |

### Data Size Guidelines:
| Dataset Size | Recommended Start |
|-------------|-------------------|
| < 1K samples | KNN, SVM, Logistic Regression |
| 1K - 100K | SVM, Decision Tree, all of above |
| 100K+ | Logistic Regression, Naive Bayes (SVM too slow) |
| Text data | Naive Bayes, SVM with linear kernel |

---
## 🧠 4. Metrics Beyond Accuracy

| Metric | Formula | Best For |
|--------|---------|----------|
| **Accuracy** | Correct / Total | Balanced classes |
| **Precision** | TP / (TP + FP) | When false positives are costly |
| **Recall** | TP / (TP + FN) | When false negatives are costly |
| **F1-Score** | 2·P·R / (P+R) | Imbalanced classes |
| **ROC-AUC** | Area under ROC | Probability ranking quality |

### ⚠️ Accuracy Paradox
With 95% negative / 5% positive data, a model predicting "all negative" gets 95% accuracy but is **useless**. Always check **F1 or AUC** for imbalanced data.
