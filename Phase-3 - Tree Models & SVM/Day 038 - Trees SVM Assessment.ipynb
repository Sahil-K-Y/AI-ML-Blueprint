## 🧠 1. Phase 3 Algorithm Summary

| Algorithm | Type | Key Idea | Best For | Avoid When |
|-----------|------|----------|----------|------------|
| **Decision Tree** | Supervised | Recursive splitting on features | Interpretable rules, mixed types | Unstable, overfits |
| **SVM** | Supervised | Maximum margin hyperplane | Medium data, non-linear (kernel) | Large datasets (slow) |
| **Naive Bayes** | Supervised | Bayes theorem + independence | Text, high-dim, fast baseline | Feature interactions |
| **KNN** | Supervised | Distance to k nearest points | Small data, local patterns | High-dim, large data |
| **Voting** | Ensemble | Combine diverse models | Quick ensemble boost | All models are similar |
| **Bagging** | Ensemble | Bootstrap + aggregate | Reduce variance of unstable models | High-bias models |
| **Random Forest** | Ensemble | Bagging + random features | General purpose, robust | Need interpretability |

## 🧠 2. Hyperparameter Quick Reference

### Decision Tree
| Param | Default | Key Effect |
|-------|---------|------------|
| `max_depth` | None | Controls overfitting |
| `min_samples_split` | 2 | Minimum to split a node |
| `criterion` | 'gini' | Split quality (gini/entropy) |
| `ccp_alpha` | 0.0 | Cost-complexity pruning |

### SVM
| Param | Default | Key Effect |
|-------|---------|------------|
| `C` | 1.0 | Regularization (low=smooth, high=complex) |
| `kernel` | 'rbf' | Decision boundary shape |
| `gamma` | 'scale' | RBF influence radius |

### KNN
| Param | Default | Key Effect |
|-------|---------|------------|
| `n_neighbors` | 5 | Smoothness of boundary |
| `weights` | 'uniform' | uniform vs distance-weighted |
| `metric` | 'minkowski' | Distance calculation |

### Random Forest
| Param | Default | Key Effect |
|-------|---------|------------|
| `n_estimators` | 100 | Number of trees |
| `max_features` | 'sqrt' | Feature randomness per split |
| `max_depth` | None | Tree depth limit |

---
## 🧠 3. Feature Importance Methods

| Method | Type | Pros | Cons |
|--------|------|------|------|
| **MDI (Impurity)** | Tree-specific | Fast, built-in | Biased toward high-cardinality |
| **Permutation** | Model-agnostic | Unbiased, on test data | Slower, affected by correlated features |
| **Drop-column** | Model-agnostic | Gold standard accuracy | Very slow (retrain per feature) |
