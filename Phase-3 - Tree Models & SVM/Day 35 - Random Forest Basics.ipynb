## 🧠 1. How Random Forest Works

### Two Sources of Randomness:

**1. Bootstrap Sampling (from Bagging)**
- Each tree gets a **random sample with replacement** from training data
- ~63.2% unique samples per tree, rest are OOB

**2. Random Feature Selection (NEW in RF)**
- At each split, only a **random subset of features** is considered
- For classification: typically `sqrt(n_features)` features
- For regression: typically `n_features / 3` features
- This **decorrelates trees**, making the ensemble more effective

### Why Random Feature Selection Matters:
Without it, if one feature is very strong, **all trees will split on it first** → trees become correlated → averaging correlated predictors gives less variance reduction.

With random features, each tree is **forced to explore different paths** → more diversity → better ensemble.

## 🧠 2. Random Forest vs Single Decision Tree

| Property | Decision Tree | Random Forest |
|----------|--------------|---------------|
| **Bias** | Low (can fit any pattern) | Low (each tree is deep) |
| **Variance** | HIGH (unstable) | LOW (averaged across trees) |
| **Overfitting** | Very prone | Much more robust |
| **Interpretability** | Easy (single tree) | Harder (forest of trees) |
| **Training Speed** | Fast | Slower (many trees) |
| **Prediction Speed** | Fast | Slower (aggregate predictions) |
| **Feature Importance** | Available | Better (averaged) |

---
## 🧠 3. Key Hyperparameters

| Parameter | Default | Description | Effect |
|-----------|---------|-------------|--------|
| `n_estimators` | 100 | Number of trees | More = better (diminishing returns after ~200) |
| `max_features` | 'sqrt' | Features per split | Lower = more diversity = less variance |
| `max_depth` | None | Tree depth limit | None = fully grown trees (higher variance per tree) |
| `min_samples_split` | 2 | Min samples to split | Higher = simpler trees |
| `min_samples_leaf` | 1 | Min samples in leaf | Higher = smoother predictions |
| `bootstrap` | True | Use bootstrap sampling | False = use full dataset (pasting) |
| `oob_score` | False | Compute OOB score | Free validation estimate |
| `n_jobs` | None | Parallel cores | -1 = use all cores |
| `criterion` | 'gini' | Split quality measure | 'gini' or 'entropy' |

### Rules of Thumb:
- **n_estimators:** Start with 100, increase to 500 if needed. More trees never hurts (only cost = time)
- **max_features:** 'sqrt' for classification, 'log2' or n/3 for regression
- **max_depth:** Leave None for most cases; limit only if overfitting severely

---
## 🧠 4. OOB Score — Free Cross-Validation

- Each tree's OOB samples (~36.8%) serve as its **validation set**
- OOB score ≈ cross-validation score, but **computationally free**
- Set `oob_score=True` and access via `rf.oob_score_`
- Use this to monitor performance during hyperparameter tuning
