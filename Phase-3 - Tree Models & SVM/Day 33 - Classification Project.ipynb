## 📊 Dataset: Online Shoppers Purchasing Intention

This dataset contains **12,330 sessions** from an online shopping website. The goal is to predict whether a visitor will **generate revenue** (make a purchase) or not.

### Features Include:
- **Numerical:** Administrative, Informational, ProductRelated page durations, BounceRates, ExitRates, PageValues
- **Categorical:** Month, OperatingSystems, Browser, Region, TrafficType, VisitorType, Weekend
- **Target:** `Revenue` (True/False — did the visitor purchase?)

**12,330 samples, 17 features, binary classification (imbalanced: ~84% No Purchase)**

**Why this dataset?** Real e-commerce data with **mixed feature types**, **class imbalance**, and **meaningful business impact** — predicting customer purchase intent.
