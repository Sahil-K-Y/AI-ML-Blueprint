# 🧠 Day 004 - Data Leakage & Cross Validation Deep Dive
### **Goal:** Demonstrate and prevent data leakage, and implement K-Fold and Stratified K-Fold cross validation.

--- 
### 🧪 Step 1: Demonstrate Data Leakage (The Wrong Way)
- **Goal:** Scale synthetic dataset *before* performing train-test split.
- **Explanation:** Fitting a scaler on all data leaks test-set properties (mean, std) into the train fold, leading to overly optimistic results.

--- 
### 🛡️ Step 2: Prevent Data Leakage using Pipelines
- **Goal:** Split raw data first, and then wrap Scaling + Logistic Regression inside a `Pipeline`.
- **Explanation:** Pipelines ensure scaling parameter estimation (`fit`) only runs on the train subset, preventing leakage.

--- 
### 🔄 Step 3: Implement Stratified K-Fold Cross Validation
- **Goal:** Generate an imbalanced target dataset, and compare normal K-Fold vs Stratified K-Fold class distributions across folds.

--- 
### 📊 Step 4: Compare CV Scores with Loop-level Leakage
- **Goal:** Compare scores when scaling happens before CV vs when scaling runs inside CV loop folds, showing leakage inflation.
