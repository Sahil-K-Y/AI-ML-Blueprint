### 📐 Step 1: Maximum Likelihood Estimation (MLE) Calculation
- **Goal:** Implement manual numpy functions to compute the log-likelihood and cross-entropy loss.
- **Explanation:** MLE finds model parameters that maximize likelihood of actual labels. Log Loss is derived directly by minimizing negative log-likelihood.

--- 
### 🧮 Step 2: Implement Softmax Function from Scratch
- **Goal:** Write a numpy function to calculate Softmax scores for 1D or 2D logit matrices: $P(y=k|x) = \frac{e^{z_k}}{\sum_j e^{z_j}}$.
- **Explanation:** Softmax converts any range of real numbers (logits) into a probability distribution summing to 1.
