# 🚢 Day 002 - [WEEKEND PROJECT 01] Titanic Survival Classifier (Part 2)
### **Goal:** Build the estimator pipeline, perform hyperparameter tuning, evaluate, and optimize decision threshold.

--- 
### ⚙️ Step 1: Reconstruct Pipeline Preprocessor
- **Goal:** Re-import raw data, apply title & family size feature engineering, split train/test, and reconstruct the preprocessing `ColumnTransformer` from Day 001.

--- 
### 🏗️ Step 2: Full Estimator Pipeline
- **Goal:** Combine the preprocessing transformer and a `LogisticRegression` classifier into a single `Pipeline` object.
- **Explanation:** Wrapping both steps prevents separate transform commands during inference, making predictions simpler.

--- 
### 🔍 Step 3: Hyperparameter Tuning with GridSearchCV
- **Goal:** Define a grid for `C`, `penalty`, and `solver` parameters, and run `GridSearchCV` on the training set.
- **Explanation:** Logistic Regression performance changes depending on penalty (`l1`/`l2`) and regularization strength (`C`). Solve for the best setup using 5-fold cross-validation.

--- 
### 📊 Step 4: Model Evaluation
- **Goal:** Run predictions on test set. Print classification report (precision, recall, F1), display confusion matrix, and compute ROC-AUC score.
- **Explanation:** Analyze errors (false positives vs false negatives) to see where the linear classifier fails.

--- 
### 🎯 Step 5: Decision Threshold Optimization
- **Goal:** Obtain class probabilities via `predict_proba`. Loop through thresholds (0.1 to 0.9) to find the cut-off that maximizes the F1-Score.
- **Explanation:** By default, classification uses `0.5`. Adjusting this boundary improves predictions on imbalanced metrics.
