```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def sigmoid(z):
    return 1 / (1 + np.exp(-z))
```

```python
x = np.linspace(-10, 10, 200)
y = sigmoid(x)

plt.figure(figsize=(8,5))
plt.plot(x, y)
plt.title("Sigmoid Function")
plt.xlabel("z")
plt.ylabel("Sigmoid(z)")
plt.grid(True)
plt.show()
```

```python
values = np.array([-5, -2, 0, 2, 5])

print("Input :", values)
print("Output:", sigmoid(values))
```

```python
def log_loss(y_true, y_pred):

    epsilon = 1e-15

    y_pred = np.clip(y_pred, epsilon, 1 - epsilon)

    loss = -(y_true*np.log(y_pred) +
             (1-y_true)*np.log(1-y_pred))

    return np.mean(loss)
```

```python
y_true = np.array([1,0,1,1])

y_pred = np.array([0.9,0.2,0.8,0.7])

loss = log_loss(y_true,y_pred)

print("Log Loss =",loss)
```

```python
y_true = np.array([1,0,1,1])

y_pred = np.array([0.2,0.8,0.3,0.1])

loss = log_loss(y_true,y_pred)

print(loss)
```

```python
good_pred = np.array([0.95,0.05,0.90,0.85])

bad_pred = np.array([0.10,0.80,0.20,0.30])

print("Good Prediction Loss :",log_loss(y_true,good_pred))

print("Bad Prediction Loss  :",log_loss(y_true,bad_pred))
```

```python
probabilities = np.array([0.1,0.3,0.49,0.51,0.75,0.9])

prediction = (probabilities >= 0.5).astype(int)

print("Probability :",probabilities)
print("Prediction  :",prediction)
```

```python
p = np.linspace(0.001,0.999,500)

loss_when_true_1 = -np.log(p)

loss_when_true_0 = -np.log(1-p)

plt.figure(figsize=(8,5))

plt.plot(p,loss_when_true_1,label="Actual = 1")

plt.plot(p,loss_when_true_0,label="Actual = 0")

plt.xlabel("Predicted Probability")

plt.ylabel("Loss")

plt.title("Binary Cross Entropy Loss")

plt.legend()

plt.grid(True)

plt.show()
```
