# 🔧 Day 005 - Regularization, Cost-Sensitive Learning & Calibration
### **Goal:** Analyze L1/L2 penalties, class weight tuning, and probability calibration curves.

--- 
### ⚖️ Step 1: L1 (Lasso) vs L2 (Ridge) Coefficients Sparsity
- **Goal:** Fit Logistic Regression models with L1 and L2 penalties on high-dimensional data.
- **Explanation:** Plot coefficients to observe how L1 regularization sets weights to exactly zero, acting as a feature selector.

--- 
### 🎛️ Step 2: Cost-Sensitive Logistic Regression
- **Goal:** Compare standard classifier performance with a weighted classifier using `class_weight='balanced'` on imbalanced data.
- **Explanation:** Balancing weights penalizes minority-class errors proportionately, improving recall on rare labels.

--- 
### 📈 Step 3: Probability Calibration Curves
- **Goal:** Generate a calibration curve using `calibration_curve`. Calibrate model probabilities using `CalibratedClassifierCV`.
- **Explanation:** Ensure that predicted probabilities correspond closely to actual class frequencies.
