# 📅 Day 031: Pipeline Verification with Pytest 🧪

Welcome to Day 031! Today's goal is to learn the engineering habit of **Unit Testing**. In professional ML engineering, we don't just run code in notebooks and inspect it manually; we write automated tests to make sure our pipelines are robust and bug-free.

---

### Why do we test ML pipelines?
1. **Silent Failures:** If your preprocessing fills null values with the wrong median or mode, python won't crash. Your model will just train on bad data and give poor predictions. Tests catch this.
2. **Feature Drift/Schema Changes:** If the incoming CSV column names change, your pipeline might break or drop columns. Tests verify column consistency.
3. **Code Safety:** If you refactor your preprocessing function in `src/pipeline.py`, tests run in 1 second to confirm you didn't break anything.

### 💡 Key Takeaways
1. **Fixtures:** We use `@pytest.fixture` to define mock datasets that can be reused across multiple test functions.
2. **Assertions:** We use `assert` to verify that our code's actual output matches our expected output.
3. **Exception Testing:** We use `with pytest.raises(...)` to make sure our code handles errors (like missing files) gracefully.
