### Goal: Perform extreme case verification and integration testing on the housing prediction app.

```python
import joblib
import numpy as np

model = joblib.load('artifacts/model.joblib')

print('--- Extreme Value Testing ---')
# 1. Median Income is extremely high
high_income = np.array([[15.0, 20.0, 5.0, 1.0, 1000.0, 3.0, 34.0, -118.0]])
print(f'Extreme High Income Price: ${model.predict(high_income)[0]*100000:,.2f}')

# 2. House Age is brand new
new_house = np.array([[3.0, 0.0, 5.0, 1.0, 1000.0, 3.0, 34.0, -118.0]])
print(f'Brand New House Price: ${model.predict(new_house)[0]*100000:,.2f}')
```
