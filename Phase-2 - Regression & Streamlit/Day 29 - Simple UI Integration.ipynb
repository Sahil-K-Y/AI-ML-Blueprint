Here, we demonstrate loading the serialized model and executing predictions dynamically inside a Streamlit application.

```python
import joblib
import numpy as np

# Load model pipeline
model = joblib.load('artifacts/model.joblib')

# Test inputs
test_input = np.array([[3.0, 20.0, 5.0, 1.0, 1000.0, 3.0, 34.0, -118.0]])
pred = model.predict(test_input)[0]
print(f'Prediction test complete. Value: ${pred * 100000:,.2f}')
```
